```python
import numpy as np
import pandas as pd
```

```python
new_movies=pd.read_csv('./cleaned_csv.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000,stop_words='english')
vectors=vectorizer.fit_transform(new_movies['keys']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
np.save('array_data.npy', similarity)
```

```python
def get_top_15_indexes(name):
    num = new_movies[new_movies['normalised_text'] == name].index
    if num.size == 0:
        num = new_movies[new_movies['title'] == name].index
    if num.size != 0:
        sorted_indices_desc = matrix[num].argsort()[0][::-1]
        return sorted_indices_desc
    else:
        return np.array([])
```

```python
for index in get_top_15_indexes('avatar') :
    movie_title = new_movies.iloc[index]['title']
    print(movie_title)
```
